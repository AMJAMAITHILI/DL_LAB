```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adagrad, RMSprop, Adadelta, Adam

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

model = Sequential([
    Dense(4, activation='relu', input_shape=(2,)),
    Dense(1, activation='sigmoid')
])

# 1 Batch GD
model.compile(optimizer=SGD(learning_rate=0.1), loss='binary_crossentropy')
model.fit(X,y,epochs=10,batch_size=4)
print(model.predict(X))

# 2 Stochastic GD
model.compile(optimizer=SGD(learning_rate=0.1), loss='binary_crossentropy')
model.fit(X,y,epochs=10,batch_size=1,verbose=0)
print(model.predict(X))

# 3 Mini Batch GD
model.compile(optimizer=SGD(learning_rate=0.1), loss='binary_crossentropy')
model.fit(X,y,epochs=10,batch_size=2,verbose=0)
print(model.predict(X))

# SGD with Momentum
model.compile(optimizer=SGD(learning_rate=0.1, momentum=0.9), loss='binary_crossentropy')
model.fit(X,y,epochs=10,verbose=0)
print(model.predict(X))

# SGD with Nesterov
model.compile(optimizer=SGD(learning_rate=0.1, momentum=0.9, nesterov=True), loss='binary_crossentropy')
model.fit(X,y,epochs=10,verbose=0)
print(model.predict(X))

# Adagrad
model.compile(optimizer=Adagrad(learning_rate=0.1, epsilon=1e-7), loss='binary_crossentropy')
model.fit(X,y,epochs=10,verbose=0)
print(model.predict(X))

# RMSprop
model.compile(optimizer=RMSprop(learning_rate=0.001, rho=0.9), loss='binary_crossentropy')
model.fit(X,y,epochs=10,verbose=0)
print(model.predict(X))

# Adadelta
model.compile(optimizer=Adadelta(learning_rate=1.0, rho=0.95), loss='binary_crossentropy')
model.fit(X,y,epochs=10,verbose=0)
print(model.predict(X))

# Adam
model.compile(optimizer=Adam(learning_rate=0.01, beta_1=0.9, beta_2=0.999), loss='binary_crossentropy')
model.fit(X,y,epochs=10,verbose=0)
print(model.predict(X))
```

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adagrad, RMSprop, Adadelta, Adam

# Load dataset
data = load_breast_cancer()
X = data.data
y = data.target.reshape(-1,1)

# Normalize
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Model function
def create_model():
    model = Sequential([
        Dense(16, activation='relu', input_shape=(X.shape[1],)),
        Dense(8, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    return model

results = []

# 1 Batch GD
model = create_model()
model.compile(optimizer=SGD(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=10, batch_size=len(X), verbose=0)
loss, acc = model.evaluate(X, y, verbose=0)
results.append(["Batch GD", len(X), loss, acc])

# 2 Stochastic GD
model = create_model()
model.compile(optimizer=SGD(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=10, batch_size=1, verbose=0)
loss, acc = model.evaluate(X, y, verbose=0)
results.append(["SGD", 1, loss, acc])

# 3 Mini-batch GD
model = create_model()
model.compile(optimizer=SGD(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=10, batch_size=32, verbose=0)
loss, acc = model.evaluate(X, y, verbose=0)
results.append(["Mini-batch GD", 32, loss, acc])

# SGD with Momentum
model = create_model()
model.compile(optimizer=SGD(learning_rate=0.01, momentum=0.9), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=10, verbose=0)
loss, acc = model.evaluate(X, y, verbose=0)
results.append(["Momentum", 32, loss, acc])#default batch size 32

# SGD with Nesterov
model = create_model()
model.compile(optimizer=SGD(learning_rate=0.01, momentum=0.9, nesterov=True), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=10, verbose=0)
loss, acc = model.evaluate(X, y, verbose=0)
results.append(["Nesterov", 32, loss, acc])

# Adagrad
model = create_model()
model.compile(optimizer=Adagrad(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=10, verbose=0)
loss, acc = model.evaluate(X, y, verbose=0)
results.append(["Adagrad", 32, loss, acc])

# RMSprop
model = create_model()
model.compile(optimizer=RMSprop(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=10, verbose=0)
loss, acc = model.evaluate(X, y, verbose=0)
results.append(["RMSprop", 32, loss, acc])

# Adadelta
model = create_model()
model.compile(optimizer=Adadelta(learning_rate=1.0), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=10, verbose=0)
loss, acc = model.evaluate(X, y, verbose=0)
results.append(["Adadelta", 32, loss, acc])

# Adam
model = create_model()
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y, epochs=10, verbose=0)
loss, acc = model.evaluate(X, y, verbose=0)
results.append(["Adam", 32, loss, acc])

# 🔹 Print Comparison Table
print("\nComparison Table:\n")
print("{:<15} {:<10} {:<15} {:<10}".format("Optimizer", "Batch", "Loss", "Accuracy"))
print("-"*55)

for r in results:
    print("{:<15} {:<10} {:<15.4f} {:<10.4f}".format(r[0], r[1], r[2], r[3]))
```
