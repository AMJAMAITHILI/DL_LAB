```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adagrad, RMSprop, Adadelta, Adam

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

model = Sequential([
    Dense(4, activation='relu', input_shape=(2,)),
    Dense(1, activation='sigmoid')
])

# 1 Batch GD
model.compile(optimizer=SGD(learning_rate=0.1), loss='binary_crossentropy')
model.fit(X,y,epochs=10,batch_size=4)
print(model.predict(X))

# 2 Stochastic GD
model.compile(optimizer=SGD(learning_rate=0.1), loss='binary_crossentropy')
model.fit(X,y,epochs=10,batch_size=1,verbose=0)
print(model.predict(X))

# 3 Mini Batch GD
model.compile(optimizer=SGD(learning_rate=0.1), loss='binary_crossentropy')
model.fit(X,y,epochs=10,batch_size=2,verbose=0)
print(model.predict(X))

# SGD with Momentum
model.compile(optimizer=SGD(learning_rate=0.1, momentum=0.9), loss='binary_crossentropy')
model.fit(X,y,epochs=10,verbose=0)
print(model.predict(X))

# SGD with Nesterov
model.compile(optimizer=SGD(learning_rate=0.1, momentum=0.9, nesterov=True), loss='binary_crossentropy')
model.fit(X,y,epochs=10,verbose=0)
print(model.predict(X))

# Adagrad
model.compile(optimizer=Adagrad(learning_rate=0.1, epsilon=1e-7), loss='binary_crossentropy')
model.fit(X,y,epochs=10,verbose=0)
print(model.predict(X))

# RMSprop
model.compile(optimizer=RMSprop(learning_rate=0.001, rho=0.9), loss='binary_crossentropy')
model.fit(X,y,epochs=10,verbose=0)
print(model.predict(X))

# Adadelta
model.compile(optimizer=Adadelta(learning_rate=1.0, rho=0.95), loss='binary_crossentropy')
model.fit(X,y,epochs=10,verbose=0)
print(model.predict(X))

# Adam
model.compile(optimizer=Adam(learning_rate=0.01, beta_1=0.9, beta_2=0.999), loss='binary_crossentropy')
model.fit(X,y,epochs=10,verbose=0)
print(model.predict(X))
```
