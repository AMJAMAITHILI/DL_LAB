**LeNet**

```python
# Import libraries
from keras import datasets
from keras.models import Sequential
from keras.layers import Conv2D, AveragePooling2D, Flatten, Dense
from keras.utils import to_categorical

# Load CIFAR-10 dataset (color images: 32x32x3)
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()

# Normalize (0–255 -> 0–1)
x_train = x_train / 255.0
x_test = x_test / 255.0

# One-hot encode labels (10 classes)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build LeNet-5 style model
model = Sequential()

# Conv layer (learn basic features)
model.add(Conv2D(6, (5,5), activation='tanh', input_shape=(32,32,3)))

# Downsample (reduce size)
model.add(AveragePooling2D((2,2)))

# More feature learning
model.add(Conv2D(16, (5,5), activation='tanh'))

# Downsample again
model.add(AveragePooling2D((2,2)))

# Convert to 1D
model.add(Flatten())

# Fully connected layers
model.add(Dense(120, activation='tanh'))
model.add(Dense(84, activation='tanh'))

# Output layer (10 classes)
model.add(Dense(10, activation='softmax'))

# Compile model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train model
model.fit(x_train, y_train,
          epochs=10,
          batch_size=128,
          validation_split=0.2)

# Evaluate
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test accuracy:", test_acc)
```

 **ReLU instead of tanh → faster + stronger learning**

**MaxPooling → keeps strongest features**

**BatchNormalization → stable training**

**Dropout → prevents overfitting**

**Deeper network → learns complex patterns (important for CIFAR-10)**

```python
# Import libraries
from keras import datasets
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from keras.utils import to_categorical

# Load CIFAR-10 (32x32 color images)
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()

# Normalize (make pixel values 0–1)
x_train = x_train / 255.0
x_test = x_test / 255.0

# One-hot encode labels (10 classes)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build model
model = Sequential()

# ---- Block 1 (learn basic patterns like edges) ----
model.add(Conv2D(32, (3,3), activation='relu', padding='same', input_shape=(32,32,3)))
model.add(BatchNormalization())          # stabilize learning
model.add(Conv2D(32, (3,3), activation='relu', padding='same'))
model.add(MaxPooling2D((2,2)))           # shrink image
model.add(Dropout(0.25))                 # reduce overfitting

# ---- Block 2 (learn more complex features) ----
model.add(Conv2D(64, (3,3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(Conv2D(64, (3,3), activation='relu', padding='same'))
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.25))

# ---- Block 3 (deep features) ----
model.add(Conv2D(128, (3,3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))
model.add(Dropout(0.25))

# Flatten (convert 2D → 1D)
model.add(Flatten())

# Fully connected layers (decision making)
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))

# Output layer (10 classes)
model.add(Dense(10, activation='softmax'))

# Compile model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train model
model.fit(x_train, y_train,
          epochs=20,
          batch_size=128,
          validation_split=0.2)

# Evaluate on test data
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test accuracy:", test_acc)
```

**GoogLeNet**

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, concatenate, GlobalAveragePooling2D
from tensorflow.keras.models import Model

def inception_module(x):
    conv1 = Conv2D(64, (1,1), activation='relu', padding='same')(x)

    conv3 = Conv2D(128, (3,3), activation='relu', padding='same')(x)

    conv5 = Conv2D(32, (5,5), activation='relu', padding='same')(x)

    pool = MaxPooling2D((3,3), strides=1, padding='same')(x)

    return concatenate([conv1, conv3, conv5, pool])

input_layer = Input(shape=(32,32,3))

x = inception_module(input_layer)
x = inception_module(x)

x = GlobalAveragePooling2D()(x)
output = Dense(10, activation='softmax')(x)

model = Model(inputs=input_layer, outputs=output)

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
model.fit(x_train, y_train, epochs=10, batch_size=64)
```

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, concatenate

def fast_inception(x):
    conv1 = Conv2D(32, (1,1), activation='relu', padding='same')(x)

    conv3 = Conv2D(32, (1,1), activation='relu', padding='same')(x)
    conv3 = Conv2D(64, (3,3), activation='relu', padding='same')(conv3)

    conv5 = Conv2D(16, (1,1), activation='relu', padding='same')(x)
    conv5 = Conv2D(32, (3,3), activation='relu', padding='same')(conv5)
    conv5 = Conv2D(32, (3,3), activation='relu', padding='same')(conv5)

    pool = MaxPooling2D((3,3), strides=1, padding='same')(x)
    pool = Conv2D(16, (1,1), activation='relu', padding='same')(pool)

    return concatenate([conv1, conv3, conv5, pool])

input_layer = Input(shape=(32,32,3))

x = fast_inception(input_layer)
x = MaxPooling2D((2,2))(x)

x = fast_inception(x)
x = MaxPooling2D((2,2))(x)

x = GlobalAveragePooling2D()(x)
output = Dense(10, activation='softmax')(x)

model = Model(inputs=input_layer, outputs=output)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
model.fit(x_train, y_train, epochs=10, batch_size=64)
```

**ResNet**

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, concatenate, GlobalAveragePooling2D
from tensorflow.keras.models import Model

def inception_module(x):
    conv1 = Conv2D(64, (1,1), activation='relu', padding='same')(x)

    conv3 = Conv2D(128, (3,3), activation='relu', padding='same')(x)

    conv5 = Conv2D(32, (5,5), activation='relu', padding='same')(x)

    pool = MaxPooling2D((3,3), strides=1, padding='same')(x)

    return concatenate([conv1, conv3, conv5, pool])

input_layer = Input(shape=(32,32,3))

x = inception_module(input_layer)
x = inception_module(x)

x = GlobalAveragePooling2D()(x)
output = Dense(10, activation='softmax')(x)

model = Model(inputs=input_layer, outputs=output)

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
model.fit(x_train, y_train, epochs=10, batch_size=64)
```

```python
from tensorflow.keras.layers import Input, Conv2D, Add, ReLU, GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model

def res_block(x, filters):
    shortcut = x

    x = Conv2D(filters, (3,3), padding='same', activation='relu')(x)
    x = Conv2D(filters, (3,3), padding='same')(x)

    x = Add()([x, shortcut])
    x = ReLU()(x)

    return x


input_layer = Input(shape=(32,32,3))

# Initial layer (smaller than original ResNet)
x = Conv2D(32, (3,3), padding='same', activation='relu')(input_layer)

# Residual blocks
x = res_block(x, 32)
x = res_block(x, 32)

x = Conv2D(64, (3,3), strides=2, padding='same', activation='relu')(x)
x = res_block(x, 64)

x = Conv2D(128, (3,3), strides=2, padding='same', activation='relu')(x)
x = res_block(x, 128)

# Head
x = GlobalAveragePooling2D()(x)
output = Dense(10, activation='softmax')(x)

model = Model(inputs=input_layer, outputs=output)

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
model.fit(x_train, y_train, epochs=10, batch_size=64)
```
